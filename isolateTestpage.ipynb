```python
import pandas as pnd
import matplotlib.pyplot as plt
import json

# an iPython  "magic" that enables the embedding of matplotlib output
%matplotlib inline
# make figures larger
plt.rcParams['figure.dpi'] = 100
```

```python
## load data from JSON files ##

with open ("testcase.json", "rt") as TESTFILE:
    Tdata = json.load(TESTFILE)
    
Tdata
```

```python
Tdatalist=Tdata['data']
Tdatalist
```

```python
Tdates=[dictionary['date'] for dictionary in Tdatalist ]
Tdates.sort()
Tdates
```

```python
def parse_date(Tdatestring):
    """Conver date string into pandas datetime object"""
    return pnd.to_datetime(Tdatestring, format="%Y-%m-%d")
```

```python
startdate=parse_date(Tdates[0])
enddate=parse_date(Tdates[-1])
print(startdate, ' to ', enddate)
```

```python
index = pnd.date_range(startdate, enddate, freq='D')
testcasedataframe=pnd.DataFrame(index=index, columns=['newtests', 'newcases'])
testcasedataframe
```

```python
for x in Tdatalist:
    date=parse_date(x['date'])
    for column in ['newcases', 'newtests']:
        if pnd.isna(testcasedataframe.loc[date, column]):
            value=float(x[column]) if x[column]!=None else 0.0
            testcasedataframe.loc[date, column]=value
            
testcasedataframe.fillna(0.0, inplace=True)
testcasedataframe
```

```python

testcasedataframe.plot(logy=True)
```

```python
testcasedataframe.to_pickle("testcasedataframe.pkl")
```

```python
import ipywidgets as wdg
import pandas as pd
import matplotlib.pyplot as plt
```

```python
series = wdg.SelectMultiple(
    options=['newcases', 'newtests'],
    value=['newcases', 'newtests'],
    rows=2,
    description='Stats:',
    disabled=False

)

scale=wdg.RadioButtons(
    options=['linear','log'],
    description='scale',
    disabled=False

)

controls=wdg.HBox([series,scale])

def testcase_graph(gcols, gscale):
    if gscale=='linear':
        logscale=False
    else:
        logscale=True
    ncols=len(gcols)
    if ncols>0:
        testcasedataframe[list(gcols)].plot(logy=logscale)
    else:
        print("xxxxx")
        print("yyyyy")
        
graph=wdg.interactive_output(testcase_graph, {'gcols':series, 'gscale': scale})

display(controls,graph)
```
